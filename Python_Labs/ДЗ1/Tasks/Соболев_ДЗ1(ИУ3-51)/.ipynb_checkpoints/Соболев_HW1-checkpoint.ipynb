Создайте функцию, реализующую вычисление по данной формуле:

\begin{align}
 result = 6a^3 - \frac{8b^2 }{4c} + 11
\end{align}

---

````python
def f1(a,b,c):
    result = 6*a*a*a-8*b*b/(4*c)+11
    
    return result
````

### Задание 4

Создайте функцию, принимающую на вход строку и возвращающую описание длины, исходя из условий:

|длина|описание|
|-----|--------|
| меньше 8 | short | 
| от 8 до 12 | semi short|
| от 13 до 17| semi long |
| больше 17 | long|

---

````python
def f4(string):
    result = []
    if len(string) < 8:
        result.append('short')
    elif 8 <= len(string) <= 12:
                result.append('semi short')
    elif 13 <= len(string) <= 17:
                    result.append('semi long')
    else: result.append('long')
    
    return result
````

### Задание 5

Создайте функцию, принимающую на вход список слов и возвращающую список, состоящий только из слов, написанных заглавными буквами.

---

````python
def f5(lst):    
    result = []
    for i in range(0, len(lst)):
        if lst[i].isupper():
            result.append(lst[i])
            
    return result
````

### Задание 6

Создайте функцию, генерирующую список четных чисел в диапазоне $[min, max]$. Не нужно обрабатывать случай, в котором $min$ и $max$ не являются целыми.

---

````python
def f6(minimum, maximum):
        
    if minimum > maximum:
        i = maximum
        maximum = minimum
        minimum = i
        
    lst = list(range(minimum, maximum + 1))
        
    return lst
````

---

### Задание 8

Создайте функцию, принимающую на вход строку, и возвращающую среднюю длину слова в строке с точностью до одного знака после запятой.

---
*Сложность: 3*

````python
def f8(string):
    words = string.split(' ')
    summa = 0.0
    result = []
    for i in words:
        summa += len(i)
    return round(summa / len(words),1)
````

### Задание 9

Создайте функцию, принимающую на вход два списка одинаковой длины и возвращающую результат попарного сложения, если это числа, и конкатенации через пробел, если это строки (или число и строка). *(Результат - список такой же длины)* Почитайте про кортежи и `zip`

---

````python
def f9(l1, l2):        
    return [x+y  if isinstance(x, int) and isinstance(y, int) else 
"".join([str(x),str(y)]) for (x, y) in set(zip(l1,l2))]
````

### Задание 10

Создайте функцию, генерирующую словарь
```python
{'name': 'Jack Smith',
 'hobbies': ['Python', 'football'],
 'age': 20}
```
получая на вход переменные:
```python
first_name  = 'Jack'

programming = 'Python'
age         = 20
```

---

### Задание 11

Создайте функцию, принимающую на вход два словаря и возвращающую словарь, совмещающий исходные словари.

```python
{'color':'black',  and  {'color':'red', ->  {'color':['black', 'red'],
 'shape':'circle'}       'size':3}      ->   'shape':'circle', 'size':3}
```
Список не должен состоять из одинаковых элементов, одного элемента или иметь другой список в качестве аргумента, и должен быть отсортирован по возрастанию.

---

````python
def f11(d1, d2): 
    sdict={} 
    for d in d1,d2: 
        for k, v in d.items(): 
            sdict.setdefault(k,[]).append(v) 
    for k, v in sdict.items(): 
        if len(v)==1: 
            sdict[k]=v[0] 
        else: 
            sdict[k]=sorted(v) 
    return sdict
````

## Легенда о ромуле и Реме
<span style="color:red">Четырнадцатый по счёту царь Альба-Лонги Нумитор был свергнут своим братом Амулием. Амулий не хотел, чтобы дети Нумитора помешали его честолюбивым замыслам: сын Нумитора пропал во время охоты, а Рею Сильвию заставили стать весталкой. На четвёртый год служения к ней в священной роще явился бог Марс, от которого Рея Сильвия и родила двух братьев. Разгневанный Амулий приказал положить младенцев в корзину и бросить в реку Тибр. Однако корзину прибило к берегу у подошвы Палатинского холма, где их вскормила волчица, а заботы матери заменили прилетевшие дятел и чибис. Впоследствии все эти животные стали священными для Рима. <big>Затем братьев подобрал царский пастух Фаустул.</big> Жена его, Акка Ларенция, ещё не утешившаяся после смерти своего ребёнка, приняла близнецов на своё попечение. Когда Ромул и Рем выросли, они вернулись в Альба-Лонгу, где узнали тайну своего происхождения. Они убили Амулия и восстановили на троне своего деда Нумитора.</span>

Древнеримский бог войны Марс и Рея Сильвия Рубенс П. П.
Спустя четыре года, по воле своего деда, Ромул и Рем отправились к Тибру искать место для основания новой колонии Альба-Лонга. По легенде, Рем выбрал низменность между Палатинским и Капитолийским холмами, но Ромул настаивал на том, чтобы основать город на Палатинском холме. Обращение к знамениям не помогло, вспыхнула ссора, в ходе которой Ромул убил своего брата. Раскаявшись в убийстве Рема, Ромул основал город, которому дал своё имя (лат. Roma), и стал его царём. Датой основания города считается 21 апреля 753 до н. э. когда вокруг Палатинского холма плугом была проведена первая борозда. По средневековой легенде, сыном Рема — Сением был основан город Сиена.

### Латины и этруски
<small>С середины II тыс. до н. э.</small> в нижнем течении Тибра расселялись латино-сикульские племена, ветвь италиков, пришедших на Апеннинский полуостров из придунайских областей в начале II тыс. до н. э. Латины обосновались на холмах Палатин и Велия, соседние холмы заняли сабины. После объединения нескольких латинских и сабинских поселков в середине VIII века до н. э. на Капитолийском холме была выстроена общая для всех крепость — Рим.

Этруски были древними племенами, населявшими в первом тысячелетии до н. э. северо-запад Апеннинского полуострова (область — древняя Этрурия, современная Тоскана) между реками Арно и Тибр и создавшими развитую цивилизацию, предшествовавшую римской и оказавшую на неё большое влияние. Римская культура унаследовала многие обычаи и традиции из культуры этрусков. Примерно во II веке до н. э., вследствие ассимиляции со стороны Рима, этрусская цивилизация прекратила своё существование.

### Начало Рима
<span style="color:green">С целью увеличить население Рима на первых стадиях его развития, Ромул предоставил пришельцам права, свободы и гражданство наравне с первыми поселенцами, для которых он отвёл земли Капитолийского холма. Благодаря этому в город начали стекаться беглые рабы, изгнанники и просто искатели приключений из других городов и стран.

В Риме также не хватало женского населения — соседние народы справедливо считали постыдным для себя вступление в родственные союзы с толпой бродяг, как они называли в то время римлян. Тогда Ромул придумал торжественный праздник — Консуалии, с играми, борьбой и разного рода гимнастическими и кавалерийскими упражнениями. На праздник съехались многие соседи римлян, в том числе сабиняне (сабины). В минуту, когда зрители и, в особенности, зрительницы были увлечены ходом игры, по условному знаку многочисленная толпа римлян с мечами и копьями в руках набросилась на безоружных гостей. В сумятице и давке римляне захватили женщин, сам Ромул взял себе в жёны сабинянку Герсилию. Свадьба с ритуалом  похищения невесты с тех пор стала ~~египетским~~ римским обычаем.</green>
